## 作业二参考代码

```python
#作业二参考代码
lst=[]
'''
输入txt：目标文本文件
计算：目标文件每一行有（股票代码、开始时间、结束时间），用每一行这些信息计算股票收盘序列
返回lst：目标文件中所有的股票收盘序列
'''
txt=open('./train_true.txt').readlines()
for i in txt:
    code=i[:6]
    if code[0]=='6':
        code+='.XSHG'
    else:
        code+='.XSHE'
    start_date=i[7:17]
    end_date=i[18:28]
    df=pd.read_csv(open('./Python(1)_stock_作业/csvfile.csv'),index_col=0)
    #列表方法index
    try:
        st=list(df.index).index(start_date)
    except:
         continue
    try:
        ed=list(df.index).index(end_date)+len(df)//4
        close=df.iloc[st:ed,][code]
    except:
         continue
    
    lst.append(close)
```

## 写成函数形式批量计算r

```python
def ma(series,n):
    '''
    input:pd.Series对象，一般为股票收盘价序列
    process:计算n日移动平均
    output:与输入长度相同的pd.Series对象
    '''
    return series.rolling(n).mean()
```

\begin{array}{c}{z_{1}^{2}}\\{z_{2}^{2}}\\{z_{3}^{2}}\end{array}

```python
def get_df(objdf):
    '''
    input:pd.Series对象 一支股票的收盘价序列
    process:计算ma5、ma10指标即：5日移动平均值、10日移动平均值
    output:包含一支股票收盘序列即ma5、ma10指标的信息dataframe
    '''
    dic={}
    \left[ \begin{array}{c}{\left(z_{1}^{(2)}\right)} \\ {z_{2}^{(2)}} \\ {z_{3}^{(2)}}\end{array}\right]

    
    
    dic['close']=objdf
    dic['ma5']=ma(objdf,5)
    dic['ma10']=list(ma(objdf,10))
    objdf=pd.DataFrame(dic,index=objdf.index)
    return objdf
```

```python
def caculate_r(objdf):
    '''
    input:一只股票的信息dataframe
    process:计算该股票序列收益
    output:期末持币数、持股数、总财产
    '''
    obj_df=objdf.iloc[:]
    hold=0
    money=10000
    for index in obj_df.index:
        if obj_df['ma5'][index]>obj_df['ma10'][index]:
            hold+=money/obj_df['close'][index]
            money=0
        elif obj_df['ma5'][index]<obj_df['ma10'][index]:
            money+=hold*obj_df['close'][index]
            hold=0
    return money,hold,money+list(obj_df['close'])[-1]*hold
```
